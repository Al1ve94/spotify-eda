A concrete example is a linear regression, in which you may transform a predictor variable such that it has a linear relation with a target variable. Common variable transformations are: calculating log transformations and polynomial features, encoding a categorical variable, and scaling a variable. 


Question remains on data tranformation:

>**Feature Scaling**
Feature scalling to allow comparison of variables with different scales. It involves adjusting the scale and there are several approaches on how to do it:
>1. Standard Scalling - converts features to standard normal variables (subtracting the mean and dividing by the SD or SE), quite sensitive to outliers
>2. Min-Max scalling  - converts variables to continous variables in the range (by mapping min to 0 and maximum to 1), very sensitive to outliers
>3. Robust scalling - simmilar to min-max scalling, but instead map interquartile range (75th - 25th percentile value) to 0,1. It means that it takes values outside the interval.

# Hypthesis Testing

As discussed in the course there are two major paradigms in statistics Frequentist and Bayesian statistics. 
Definitions:
* **Test statistic**: likelihood ratio, we use it to decide whether to accept/reject Ho.
* **Rejection region**: the set of values of the test statitstic that lead to rejection of Ho.
* **Acceptance region**: set of values of the test statistic that lead to acceptance of Ho.
* **Null Distribution**: test statistic distribution when null is true.

2 Types of Possible Errors:
* Type 1 Error (False Positive): Incorrectly Rejecting Null, but it is actually true.
* Type 2 Error (False Negative): Incorrectly Accepting Null, but it is actually false.

Next, we will obtain our statistics, t-value and p-value. We will use scipy.stats library and ttest_ind() function to calculate these parameters.

````python
# Importing libraries
import scipy.stats as stats 
from scipy.stats import chi2_contingency

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm 
````

````python
# Performing significance testing
alpha=0.05
t_value1, p_value1 = stats.ttest_ind(dance_minor, dance_major)
print("t_value1 = ",t_value1, ", p_value1 = ", p_value1)
````

Some 'if/else' statements to make our conclusions about the the hypothesis. 
If the p-value is smaller than our threshold, then we have evidence against the null hypothesis of equal population means.

````python
# Printing the Results and Conclusion
if p_value1 < alpha:
    print("Results:\np_value {} is less than alpha {}\n". format (p_value1,alpha))
    print("Conclusion:\nReject the null hypothesis that there is no difference between danceability of major and minor modes.")
    
else:
    print("Results:\n p_value {} is greater than alpha {}\n ". format (p_value1,alpha))
    print("Conclusion: Fail to reject the null hypothesis that there is a difference between danceability of major and minor modes.")
````

So essentially we discovered that **there is difference** between modes of danceability. Mode interestingly is that, even though by count there are more major songs in a dataset, relative to danceability, it is in fact songs in **minor** that have a higher danceability mean and density.
